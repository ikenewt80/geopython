```python
#problem 4

#function creates shapely points for origin

def origin (row):
    x = row["from_x"]
    y = row["from_y"]
    return Point (x,y)

#function creates shapely points for destination

def destination (row):
    x = row["to_x"]
    y = row["to_y"]
    return Point (x,y)

#function makes line string of start and end points

def createLine(row):
    start = row["orig_points"]
    finish= row["dest_points"]
    return LineString([start,finish])

#function claculates the distance of the lines

def distance (row):
    line = row["lines"]
    return  line.length
   

import pandas as pd

from shapely.geometry import Point,LineString

file = pd.read_csv(r'C:\Users\admin\Documents\SCHOOL\Geopy\travelTimes_2015_Helsinki.txt', sep=';', encoding='latin1')

file ["orig_points"] = file.apply(origin,axis=1)


file ["dest_points"] = file.apply(destination,axis=1)


file["lines"] = file.apply(createLine,axis=1)

file["Distance"] = file.apply(distance,axis=1)

print(file.head(5))

#calculates the mean of the distances

mean_dist = file.loc[:,"Distance"].mean()

print ("{0:.4f}".format(mean_dist))

print(file.describe())

```
